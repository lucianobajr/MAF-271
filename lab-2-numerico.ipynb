# 1) Algoritmo do Ponto Fixo

```python
def pontofixo(phi,x0,precisao):
    err = 10
    i = 0
    print('  i     xn        erro\n')
    print('%3i    %6.3f'%(i,x0))

    x = 0

    while(err > precisao):
      i += 1
      x = phi(x0)
      err = abs(x - x0)
      print('%3i    %6.3f    %6.3f'%(i,x, err))
      x0 = x

    return x
```

## 1.1) Exemplo 1

```python
def function(x):
    return np.sqrt(np.sin(x))

x = pontofixo(function,0.9,0.002)

print("\n  ")
print("x = %0.15f"%x)
```

## 1.2) Exemplo 2

```python
def second_function(x):
    return np.sqrt(6 - x)

x = pontofixo(second_function,1.5,0.001)

print("\n  ")
print("x = %0.15f"%x)
```

## 2) Algoritmo do Teorema de lagrange

```python
class Data:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def lagrange(f: list, xi: int, n: int) -> float:
    resultado = 0.0
    for i in range(n):
        termo = f[i][1]
        for j in range(n):
            if j != i:
                termo = termo * (xi - f[j][0]) / (f[i][0] - f[j][0])
 
        resultado += termo
 
    return resultado
```

Para `x = 3`, com os valores de x = [0, 1, 2, 5] e fx = [2, 3, 12,147], temos: 


```python
f = [[0,2],[1,3],[2,12],[5,147]]

print("O valor de f(3) é :", lagrange(f, 3, 4))
```

Para `x = 4`, com os valores de x = [2, 3, 5, 8, 12] e fx = [10, 15, 25, 40, 60], temos: 


```python
f = [[2,10],[3,15],[5,25],[8,40],[12,60]]

print("O valor de f(4) é :", lagrange(f, 4, 5))
```
