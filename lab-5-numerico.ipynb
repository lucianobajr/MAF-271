# 1)Método de newton para sistemas não lineares

```python
import numpy as np
import warnings
import sympy
import math
```

Para este exemplo usaremos a função $f(x) = x^2 - 3x -2$

```python
def func(x):
    return x**2 - 3*x - 2

def dfunc(x):
    return 2*x - 3
```

```python
def newton_method_for_nonlinear_systems(f,fder,x0,eps=0.0001,maxiter=1000):
    iterator = 0
    for i in range (maxiter):
        x_f = f(x0)
        x_fder = fder(x0)
        x1 = x0 - (x_f/x_fder)
        delta = abs(x1 - x0)
        if delta <= eps:
          return x1
        x0 = x1
        print("Iteration " + str(i) + ": " + str(x0))
    return x0

solution = newton_method_for_nonlinear_systems(func,dfunc,0.5)
solution
```
