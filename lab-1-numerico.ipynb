# 1) Criação de um algoritmo para o método da Bissecção

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
```

```python
def function(x):
    return x**2 - 2

def bisseccao(f,a,b,epsilon):
    m = (a+b)/2
    fm = f(m)
    fa = f(a)
    i = 0

    print('  i     xn        erro\n\n')
    while (abs(b - a)>epsilon):
        if fa*fm>0:
            a = m
            fa = fm
        else:
            b = m

        m = (a+b)/2
        fm = f(m)
        i+=1
        print('%3i    %6.3f    %6.3f'%(i,m,abs(b-a)))

bisseccao(function,1,3,0.02)
```

# 2) Criação de um algoritmo para o método de Newton

```python
def df_calc(parameter):
  x = Symbol('x')
  func = function(x)
  return func.diff(x).subs(x,parameter) 

def df_equation():
  x = Symbol('x')
  func = function(x)
  return func.diff(x)

def newton(f,df,x0,epsilon):
  x = x0
  xn = 2*x
  fx = f(x)
  dfx = df_calc(x)
  i = 0
  erro = abs(x - xn)
  print('  i     xn        f(xn)     erro\n')
  print('%3i    %6.3f    %6.3f'%(i,x,fx))

  while(erro > epsilon):
    xn = x
    x = x - fx/dfx
    erro = abs(x - xn)
    fx = f(x)
    dfx = df_calc(x)
    i +=1
    print('%3i    %6.3f    %6.3f    %6.3f'%(i,x,fx,erro))


newton(function,df_equation,3,0.02)
```

# 3) Criação de um algoritmo para Secante

```python
def secante(f,x0,c,epsilon):
  x = x0
  fc = f(c)
  fx = f(x)
  i = 0
  erro = abs(x - c)
  print('  i     xn        f(xn)     erro\n')
  print('%3i    %6.3f    %6.3f'%(i,x,fx))

  while(erro > epsilon):
    x = x - (fx * (x - c))/(fx - fc)
    erro = abs(x - x0)
    fx = f(x)
    x0 = x
    i +=1
    print('%3i    %6.3f    %6.3f    %6.3f'%(i,x,fx,erro))

secante(function,1,2,0.02)
```
