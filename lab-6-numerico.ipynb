# 1) Regra de Trapézio

## Para este exemplo usaremos a função $f(x) = e^x - 2$

```python
def function(x):
  y = np.exp(x)-2
  return y

def trapezio(fn,a, b, n):
    h = (b - a)/n
    x = []
    
    # Definição do primeiro ponto depois de a
    first_point = a + h
    # Insere o primeiro ponto na lista geral
    x.append(first_point)
    # Insere o primeiro valor no somatório
    sum = fn(first_point)
    # Insere os valores na lista e calcula o somatório dos valores na função
    for i in range(1, n - 1, 1):
        x.append(x[i - 1] + h)
        sum += fn(x[i])
    result = (h/2) * (fn(a) + (2*(sum)) + fn(b))
    print('resultado =', result)
```

```python
trapezio(function,0,2,10)
```

# 2) Regra de Simpson

## Para este exemplo usaremos a função $f(x) = e^x - 2$

```python
def simpson(fn,a, b, n):
    h = (b - a)/n
    x = []

    # Definição do primeiro primeiro ponto depois de a
    first_point = a + h
    
    x.append(round(first_point, 1))
    x_impar = []
    x_par = []
    soma_par = 0
    soma_impar = 0
    result = 0
    count = 0

    #Insere os itens na lista geral
    for i in range(1, n - 1, 1):
        x.append(x[i - 1] + h)

    #Insere os itens nas posicoes impares na lista de itens impares
    for i in range(0, n - 1, 2):
        x_impar.append(x[i])
        soma_impar += fn(x_impar[count])
        count += 1
    count = 0
    #Insere os itens nas posicoes pares na lista de itens pares
    for i in range(1, n - 1, 2):
        x_par.append(x[i])
        soma_par += fn(x_par[count])
        count += 1
    #Formula
    result = (h/3)*(fn(a) + 2*(soma_par) + 4*(soma_impar) + fn(b))
    print('resultado =', result)
```

```python
simpson(function,0, 2, 10)
```

# 3) Método de Euler

## Para esse método usaremos $ \frac{dy}{dx} = (x + y + xy)$

```python
def func( x, y ):
    return (x + y + x * y)
    
def euler( x0, y, h, x ):
    temp = 0
 
    # Iterando até o ponto em que
    # precisa de aproximação
    while x0 < x:
        temp = y
        y = y + h * func(x0, y)
        x0 = x0 + h
 
    print("Solução aproximada em x = ", x, " é ", "%.6f"% y)
```
